```python
task_description = """
Build a machine learning model that detects anomalies in data using isolation forest algorithm.
"""
```

```python
submission_text = """
This project builds an anomaly detection system using Isolation Forest.
The model detects unusual patterns in datasets.
"""
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
task_embedding = model.encode(task_description)
submission_embedding = model.encode(submission_text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(
    [task_embedding],
    [submission_embedding]
)[0][0]

print("Similarity Score:", similarity)
```

```python
if similarity > 0.75:
    print("Status: Valid")

elif similarity > 0.50:
    print("Status: Partially Valid")

else:
    print("Status: Invalid")
```
